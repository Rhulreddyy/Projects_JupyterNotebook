```python
file = pd.read_table('customers.csv', sep='\t', index_col=0)
file.head()
```

```python
file.isnull().any()
```

```python
file.fillna(file['Income'].mean(), inplace=True)
file.isnull().any()
```

```python
file['Total Spending']=file['MntWines']+file['MntFruits']+file['MntMeatProducts']+file['MntFishProducts']+file['MntSweetProducts']+file['MntGoldProds']
file.head()
```

```python
file.groupby('Education')['Total Spending'].mean()
```

```python
file.groupby(['Education','Marital_Status'])['Income'].mean()
```

```python
mask = (file['Education'] == 'Master') | (file['Education'] == 'PhD')
file[mask].sort_values(by='Total Spending', ascending=False).head(10)
```

```python
file['Dt_Customer'].dtype
```

```python
file['Dt_Customer'] = pd.to_datetime(file['Dt_Customer'], dayfirst=True)
file.head()
```

```python
file['Dt_Customer'].dtype
```

```python
file.groupby(file['Dt_Customer'].dt.year).size()
```
